```python
import numpy as np

# Исходный python-список
array_python = [2.134, 81.001, 1231.1, 2038.999, 22.421, -0.001]

# 1. Преобразование списка в массив NumPy и его отображение
arr = np.array(array_python)

# 2. Создание массива, заполненного нулями
zeros = np.zeros(6)

# 3. Создание массива, заполненного единицами
ones = np.ones(6)

# 4. Формирование последовательности чисел с равным шагом
arange = np.arange(3, 9, 1)

print(f"1. Преобразованный NumPy-массив: {arr}")
print(f"2. Массив из нулей: {zeros}")
print(f"3. Массив из единиц: {ones}")
print(f"4. Последовательность значений: {arange}")

# 5. Получение сведений о форме массива и его типе данных
print(f"5. Характеристики массива py_array → форма: {arr.shape}, тип данных: {arr.dtype}")

# 6. Доступ к элементам массива и выделение диапазона
print(f"6. Первый элемент и подмассив [1:3]: {arr[0]} ; {arr[1:3]}")

# 7. Векторное сложение двух массивов
print(f"7. Результат покомпонентного сложения arr1 и arr3: {np.add(arr, ones)}")

# 8. Векторное умножение массивов
print(f"8. Результат покомпонентного умножения arr1 и arr4: {np.multiply(arr, arange)}")

# 9. Среднее значение элементов массива
print(f"9. Среднее арифметическое элементов arr1: {arr.mean()}")
```
